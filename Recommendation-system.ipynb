```python
rating_name = df_ratings.merge(df_books,on='ISBN')
```

```python
num_rating = rating_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating
```

```python
rating_name['Book-Rating'] = pd.to_numeric(rating_name['Book-Rating'], errors='coerce')

avg_rating = (
    rating_name.groupby('Book-Title')['Book-Rating']
    .mean()
    .reset_index()
    .rename(columns={'Book-Rating': 'avg_rating'})
)

avg_rating.head()
```

```python
comb_df = num_rating.merge(avg_rating,on='Book-Title')
comb_df
```

```python
comb_df = comb_df[comb_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50) 
```

```python
comb_df = comb_df.merge(df_books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
comb_df['Image-URL-M'][0]
```

```python
x = rating_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = rating_name[rating_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df_books[df_books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data 
```

```python
recommend('1984')
```
